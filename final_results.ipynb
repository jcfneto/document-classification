```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import f1_score, confusion_matrix, ConfusionMatrixDisplay

sns.set_theme(style='white')
```

```python
results = pd.read_csv('output/results/results.csv')
classifiers = results.columns[1:].to_list()
results.head(1)
```

```python
# f1 score
f1_micro = [f1_score(results.doc_class.values, results[c].values, average='micro') for c in classifiers]
f1_macro = [f1_score(results.doc_class.values, results[c].values, average='macro') for c in classifiers]
f1 = [f1_score(results.doc_class.values, results[c].values, average=None) for c in classifiers]
```

```python
# F1 micro
sns.barplot(y=f1_micro, x=classifiers)
plt.grid(axis='y')
plt.ylabel('F1 Micro')
plt.show()
```

```python
# F1 macro
sns.barplot(y=f1_macro, x=classifiers)
plt.grid(axis='y')
plt.ylabel('F1 Macro')
plt.show()
```
